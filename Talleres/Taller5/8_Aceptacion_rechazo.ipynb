```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
from scipy import integrate
```

$$ \sqrt{Var(\bar{x})} = \frac{1}{\sqrt{N}} = Desviaci√≥n(\bar{x}) = S(\bar{x}) $$

Entonces se tiene que:

$$ N = \frac{1}{S(\bar{x})^2} $$


```python
#GRAFICO
x = np.linspace(0,1,50)
y = beta.pdf(x,2,4)
plt.plot(x,y)
plt.show()
```

```python
def integrar(a,b,er):
    """
    a: alpha
    b: beta
    c: valor inferior de la integral
    d: valor superior de la integral
    er: porcentaje de error (desviaci√≥n) que desea
    """
    n=int(1/(er/100)**2)
    x=np.random.uniform(0,1,n)
    y=np.random.uniform(0,2,n)
    It = 0
    pt = []
    for i in range (n):
        f_xi = beta.pdf(x[i],a,b)
        if y[i] < f_xi:
            It += 1
            pt.append(f_xi)
    return (It/n)*2

integrar(2,4,1)
```
